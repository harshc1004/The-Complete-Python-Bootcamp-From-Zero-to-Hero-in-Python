```python
from PIL import Image
```

```python
face = Image.open('example.jpg')
```

```python
face
```

```python
# Cropping Image

face.size
```

```python
half = 1257/2
```

```python
x = half - 200
y = half + 200
```

```python
w = 1650
h = 1257
```

```python
face.crop((x,y,w,h))
```

```python
conf = face.crop((x,y,w,h))
```

```python
face.paste(im = conf, box = (0,0))
```
